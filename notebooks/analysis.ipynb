changing columns dtypes

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.duplicated().sum()
```

Feature engineering (creating columns )

```python
df['month'] = df['Order Date'].dt.month
df['year'] = df['Order Date'].dt.year
df['month_name'] = df['Order Date'].dt.month_name
```

```python
df['profit_margin'] = df['Profit'] /df['Sales']
```

```python
df.columns=df.columns.str.lower()
```

```python
df.columns
```

```python
df.columns = df.columns.str.replace(' ','_')
```

EDA

```python
df.groupby('category')['profit'].sum().sort_values()
```

creating a summary table to see desired data for good understanding 

comparing profit to sales to see how much profit we made (or better to create profit margin column)

```python
summary = df.groupby('category')[['sales','profit']].sum()
```

```python
summary['Profit_margin'] = summary['profit'] / summary['sales']
```

Top products by sales

```python
product_summary = df.groupby('product_name')[['sales','profit']].sum()
```

```python
product_summary['profit_margin'] = product_summary['profit']/product_summary['sales']
```

```python
product_summary.sort_values('sales',ascending=False).head(10)
```

Region analysis

```python
df.groupby('region')[['sales','profit']].sum()
```

```python
region_summary = df.groupby('region')[['sales','profit']].sum()
region_summary['profit_margin'] = region_summary['profit']/region_summary['sales']
region_summary.sort_values('profit_margin',ascending=False)
```
